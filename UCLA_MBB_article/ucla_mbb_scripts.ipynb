```
# *****
# EFF by team

# load csv team data
team_data = pd.read_csv("ucla_mbb_yrs.csv")
ucla_effs = []
opp_effs = []
yrs = ["2018", "2019", "2020", "2021"]

# ucla effs
for i in range(6, -1, -2):
  team = team_data.iloc[i]
  eff = team['PTS'] + team['TRB'] + team['AST'] + team['STL'] + team['BLK'] - ((team['FGA'] - team['FG']) + (team['FTA'] - team['FT']) + team['TOV'])
  ucla_effs.append(eff)

# opp effs
for j in range(7, 0, -2):
  team = team_data.iloc[j]
  eff = team['PTS'] + team['TRB'] + team['AST'] + team['STL'] + team['BLK'] - ((team['FGA'] - team['FG']) + (team['FTA'] - team['FT']) + team['TOV'])
  opp_effs.append(eff)

# plot
plt.plot(yrs, ucla_effs, label="UCLA")
plt.plot(yrs, opp_effs, label="Opponent")

plt.xlabel('Season')
plt.ylabel('EFF')
plt.title('UCLA Overall Team EFF Over the Years')
plt.legend()
plt.show()
```

```
# *****
# UCLA vs. Opp AST/TO Ratio
yrs = ["2018", "2019", "2020", "2021"]
ucla_ratios = []
opp_ratios = []

# ucla ratios
for i in range(6, -1, -2):
  team = team_data.iloc[i]
  rat = team['AST'] / team['TOV']
  ucla_ratios.append(rat)

# opp ratios
for j in range(7, 0, -2):
  team = team_data.iloc[j]
  rat = team['AST'] / team['TOV']
  opp_ratios.append(rat)


# plot
plt.plot(yrs, ucla_ratios, label="UCLA")
plt.plot(yrs, opp_ratios, label="Opponent")

plt.xlabel('Season')
plt.ylabel('AST/TO Ratio')
plt.title('UCLA vs. Opponent AST/TO Ratio')
plt.legend()
plt.show()
```

```
# *****
# UCLA vs. Opp TRB
yrs = ["2018", "2019", "2020", "2021"]
ucla_reb = []
opp_reb = []

# ucla ratios
for i in range(6, -1, -2):
  team = team_data.iloc[i]
  ucla_reb.append(team['TRB'])

# opp ratios
for j in range(7, 0, -2):
  team = team_data.iloc[j]
  opp_reb.append(team['TRB'])


# plot
plt.plot(yrs, ucla_reb, label="UCLA")
plt.plot(yrs, opp_reb, label="Opponent")

plt.xlabel('Season')
plt.ylabel('TRB')
plt.title('UCLA vs. Opponent Rebounding')
plt.legend()
plt.show()
```

```
# *****
# UCLA PTS allowed on defense
yrs = ["2018", "2019", "2020", "2021"]
pts_allowed = []

# points allowed = opp points scored
for j in range(7, 0, -2):
  team = team_data.iloc[j]
  pts_allowed.append(team['PTS'])


# plot
plt.plot(yrs, pts_allowed, color='r')
plt.xlabel('Season')
plt.ylabel('PTS')
plt.title('Points Allowed on Defense')
plt.show()
```
